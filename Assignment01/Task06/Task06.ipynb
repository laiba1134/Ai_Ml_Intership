Task 6: House Price Prediction

Problem Statement
        House prices depend on factors like area, bedrooms, bathrooms, parking, furnishing status, and location. Manual price estimation is often inaccurate. The goal of this project is to predict house prices using machine learning.

Goal
    Build and compare two regression models:
    Linear Regression
    Gradient Boosting Regressor
Evaluate them using:
    MAE
    RMSE
    R² Score

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Step 1: Load Dataset
df = pd.read_csv("Housing.csv")

# Show first rows
print(df.head())
```

```python
# Step 2: Basic Information
print(df.info())
print(df.isnull().sum())

# Step 3: Handle Missing Values
# Fill numerical missing values with median
for col in df.select_dtypes(include=np.number).columns:
    df[col].fillna(df[col].median(), inplace=True)

# Fill categorical missing values with mode
for col in df.select_dtypes(include='object').columns:
    df[col].fillna(df[col].mode()[0], inplace=True)


# Step 4: Encode Categorical Columns
# Convert text columns into numeric columns
df = pd.get_dummies(df, drop_first=True)
```

```python
# Step 5: Define Features and Target
#  Target column is Price
X = df.drop("price", axis=1)
y = df["price"]
```

```python
# Step 6: Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Step 7: Train Test Split


X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
# Step 8: Train Models

# Linear Regression
lr = LinearRegression()
lr.fit(X_train, y_train)

# Gradient Boosting
gb = GradientBoostingRegressor()
gb.fit(X_train, y_train)
```

```python
# Step 9: Predictions
lr_pred = lr.predict(X_test)
gb_pred = gb.predict(X_test)

# Step 10: Evaluation Function
def evaluate_model(name, y_true, pred):
    mae = mean_absolute_error(y_true, pred)
    rmse = np.sqrt(mean_squared_error(y_true, pred))
    r2 = r2_score(y_true, pred)

    print(f"\n{name} Results")
    print("MAE :", mae)
    print("RMSE:", rmse)
    print("R2 Score:", r2)

# Evaluate both models
evaluate_model("Linear Regression", y_test, lr_pred)
evaluate_model("Gradient Boosting", y_test, gb_pred)

```

```python
import matplotlib.pyplot as plt
# Step 11: Visualization
plt.figure(figsize=(10,5))

plt.scatter(y_test, gb_pred, alpha=0.7)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted House Prices")

plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')
plt.show()
```

```python
# Step 12: Compare Models
results = pd.DataFrame({
    "Model": ["Linear Regression", "Gradient Boosting"],
    "MAE": [
        mean_absolute_error(y_test, lr_pred),
        mean_absolute_error(y_test, gb_pred)
    ],
    "RMSE": [
        np.sqrt(mean_squared_error(y_test, lr_pred)),
        np.sqrt(mean_squared_error(y_test, gb_pred))
    ]
})

print("\nComparison Table:")
print(results)
```

Results
    Linear Regression gave decent baseline predictions.
    Gradient Boosting performed better because it handles complex relationships between features.
Final Insights
    Important factors affecting price are area, bathrooms, location, and furnishing status.
    Lower MAE/RMSE and higher R² indicate better performance.
    Gradient Boosting is the better model for accurate house price prediction.

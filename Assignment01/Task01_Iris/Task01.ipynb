Task 1: Exploring and Visualizing a Simple Dataset

Problem Statement
  The Iris dataset contains measurements of flowers from three different species. The goal of this task is to explore the dataset to understand its structure, feature distribution, and relationships between variables using basic data analysis and visualization techniques.
Goal
  Load and inspect the Iris dataset
  Understand data structure using .shape, .info(), and .describe()
  Visualize relationships using scatter plots
  Analyze distributions using histograms
  Detect outliers using box plots

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#Step01: Loading CSV file
data = pd.read_csv('Iris.csv')
print("Shape of dataset: ",data.shape) #Print the no of(rows,columns)
print("Column names: ",data.columns) #Print the names of columns
print(data.head(12))#Print the number of rows enclosed in Bracket
data.info() #checking non-null values and understanding data types
print(data.describe(include = 'all'))
data.drop('Id', axis=1, inplace=True)
sns.scatterplot(x='SepalLengthCm', y='PetalLengthCm', hue='Species', data = data)
plt.show()
#Histogram
data.hist(figsize=(10,8))
plt.show()
#Box Plot
sns.boxplot(data=data)
plt.show()
```

  Shape, columns, head, info, and describe included
  Scatter plot implemented (feature relationship)
  Histogram used (distribution analysis)
  Box plot used (outlier detection)
